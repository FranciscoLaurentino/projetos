Desafio 18

```python
import math
print(f'{' Desafio 18 ':=^20}\n')

graus = int(input('Diga um ângulo: '))

radianos = math.radians(graus)

tan = math.tan(radianos)
cos = math.cos(radianos)
sen = math.sin(radianos)


print(f'\nTamgente: {tan:.2f}\nCosseno: {cos:.2f}\nSeno: {sen:.3f}')
```

Desafio 19

```python
from random import randint
from time import sleep
print(f'{' Desafio 19 ':=^20}\n')

cont = 0

alunos = ['Ana','Livia','Andre','Joaquim']

print(alunos)

while cont < 3:
  print('.',end='')
  sleep(1)
  cont = cont + 1
  print('.',end='')
  sleep(1)
  cont = cont + 1
  print('.')
  sleep(1)
  cont = cont + 1

sorteado = randint(0,3)

print(f'Aluno sorteado: {alunos[sorteado]}')
```
